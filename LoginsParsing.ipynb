```python
all_logins = []

for i in range(6, 40):
    book_url = 'https://www.livelib.ru/readers/rating/listview/wide/~'+str(i)
    book_html = requests.get(book_url).text
    
    soup = BeautifulSoup(book_html, 'html.parser')
    
    logins = list(map(lambda l: l.text, soup.find_all('a', class_='urow-login')))
    print(logins)
    all_logins += logins
    sleep(5)
        
    #'countries': list(map(lambda e: e.attrs['title'], soup.find_all('div', class_='flag'))),
    #'time': re.search('\d\d:\d\d', soup.find('td', class_='time').text).group(0),
    #'rating': float(soup.select_one('.rating_ball').text) # поддержка CSS-селекторов
```

```python
all_logins3 = []

headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) ' +
                         'AppleWebKit/537.36 (KHTML, like Gecko) ' +
                         'Chrome/42.0.2311.90 Safari/537.36'}

proxies = [ { 'https': '188.134.92.52:61942' },
            { 'https': '91.220.166.148:50087' },
            { 'https': '95.80.119.57:41771' } ]

import random

def random_sleep(offset=1.5, length=4):
    sleep(random.random() * length + offset)

for i in range(400, 401): #меняю вручную и проверяю, что всё выкачалось, так как иногда для рандомных страниц выдаёт пустые списки
    book_url = 'https://www.livelib.ru/readers/rating/listview/wide/~'+str(i)
    book_html = requests.get(book_url, proxies=proxies[random.randint(0,2)]).text
    len(book_html)
    
    soup = BeautifulSoup(book_html, 'html.parser')
    
    logins = list(map(lambda l: l.text, soup.find_all('a', class_='urow-login')))
    print(logins)
    all_logins3 += logins
    random_sleep()
```
