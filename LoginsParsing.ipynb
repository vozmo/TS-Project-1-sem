```python
import re
import requests
import json
from html.parser import HTMLParser

from bs4 import BeautifulSoup
from time import sleep
import random

all_logins = []
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) ' +
                         'AppleWebKit/537.36 (KHTML, like Gecko) ' +
                         'Chrome/42.0.2311.90 Safari/537.36'}
proxies = [ { 'https': '188.134.92.52:61942' },
            { 'https': '91.220.166.148:50087' },
            { 'https': '95.80.119.57:41771' },
            { 'https': '195.246.125.130:49566' },
            { 'https': '84.52.98.170:38457' }, 
            { 'https': '212.164.208.73:58992' },
            { 'https': '89.188.166.166:60629' },
            { 'https': '46.181.189.114:44960' },
            { 'https': '89.250.149.114:60981' },
            { 'https': '176.215.237.135:41529' }]
def random_sleep(offset=3, length=4):
    sleep(random.random() * length + offset)
```

```python
i = 500
while i < 1000:
    reader_url = 'https://www.livelib.ru/readers/rating/listview/wide/~'+str(i)
    print(reader_url)
    try:
        reader_html = requests.get(reader_url, headers=headers, proxies=proxies[random.randint(0, len(proxies)-1)]).text
    except:
        print('error in '+str(i))
        continue
    if len(reader_html) == 0:
        print('PUSTO in '+str(i))
        continue
    
    soup = BeautifulSoup(reader_html, 'html.parser')
    
    logins = list(map(lambda l: l.text, soup.find_all('a', class_='urow-login')))
    print(logins)
    all_logins += logins
    #random_sleep()
    i = i + 1
```
