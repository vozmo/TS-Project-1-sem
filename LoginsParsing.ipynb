```python
from bs4 import BeautifulSoup
from time import sleep
```

```python
all_logins = []

for i in range(6, 40):
    book_url = 'https://www.livelib.ru/readers/rating/listview/wide/~'+str(i)
    book_html = requests.get(book_url).text
    
    soup = BeautifulSoup(book_html, 'html.parser')
    
    logins = list(map(lambda l: l.text, soup.find_all('a', class_='urow-login')))
    print(logins)
    all_logins += logins
    sleep(5)
        
    #'countries': list(map(lambda e: e.attrs['title'], soup.find_all('div', class_='flag'))),
    #'time': re.search('\d\d:\d\d', soup.find('td', class_='time').text).group(0),
    #'rating': float(soup.select_one('.rating_ball').text) # поддержка CSS-селекторов
```
